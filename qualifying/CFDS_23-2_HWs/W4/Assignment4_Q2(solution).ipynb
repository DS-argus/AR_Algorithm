```python
from collections import deque

class Office():
    # Do not modify this code
    def __init__(self, M, N, H, employees):
        self.M = M  # 건물의 가로 길이
        self.N = N  # 건물의 세로 길이
        self.H = H  # 건물의 높이
        self.offices = [[[0] * M for _ in range(N)] for _ in range(H)]
        self.coordinates_employees = employees # 직원의 좌표를 저장하는 리스트
        self.employees()

    # Do not modify this code
    def employees(self):
        for x, y, z in self.coordinates_employees:
            if 0 <= x < self.M and 0 <= y < self.N and 0 <= z < self.H:
                self.offices[z][y][x] = 1

    def initial_stealthguard(self, coordinates_safeguards) :
        ###################################################
        # Q2-1#
        # Write your own code #
        for x,y,z in coordinates_safeguards:
            if 0 <= x < self.M and 0 <= y < self.N and 0 <= z < self.H:
                self.offices[z][y][x] = 2
        ###################################################

    def install_stealthguard(self):
        dx = [0, 0, 1, -1, 0, 0]
        dy = [1, -1, 0, 0, 0, 0]
        dz = [0, 0, 0, 0, 1, -1]

        def bfs_or_dfs(self) :
            hours_needed = 0
            ###################################################
            # Q2-2#
            # Write your own code #
            guarded_positions = deque()
            for x, y, z in self.coordinates_employees:
                    if 0 <= z < self.H and 0 <= x < self.M and 0 <= y < self.N and self.offices[z][y][x] == 2:
                        guarded_positions.append((z, x, y, 0))

            while guarded_positions:
                z, x, y, hours = guarded_positions.popleft()
                hours_needed = max(hours_needed, hours)

                for i in range(6):
                    nz, nx, ny = z + dz[i], x + dx[i], y + dy[i]
                    if 0 <= nz < self.H and 0 <= nx < self.M and 0 <= ny < self.N :
                        if self.offices[nz][ny][nx] == 1:
                            self.offices[nz][ny][nx] = 2
                            guarded_positions.append((nz, nx, ny, hours + 12))
            ###################################################
            return hours_needed


        ###################################################
        # Q2-3 #
        # Write your own code #
        hours_needed = bfs_or_dfs(self)
        days_needed = hours_needed // 24 if hours_needed % 24 == 0 else ((hours_needed // 24) + 1)
        unguarded_employees = sum(1 for z in range(self.H) for x in range(self.M) for y in range(self.N)
                                  if self.offices[z][y][x] == 1)
        ###################################################

        if unguarded_employees > 0 :
            print(f"직원 {unguarded_employees}명의 PC에 SafeGuard 프로그램을 설치할 수 없습니다.")
        else :
            print(f"{days_needed}일 후면, 모든 사무실에 SafeGuard 프로그램이 설치됩니다.")
```
