```python
#Do not modify this code
class LinkedNode():
  def __init__(self, x:int) -> None:
    self.val = x
    self.next = None
```

```python
def MakeNode(arr:list, root:LinkedNode) -> LinkedNode:
    ###########################################################
    # Q3-1 #
    # Write your own code #
    start = root
    for num in arr:
        new_node = LinkedNode(num)
        root.next = new_node
        root = root.next

    return start
    ###########################################################
```

```python
def MergeGame(arr:list) -> bool:
    root = LinkedNode(0)
    MakeNode(arr, root)
    head = root.next
    ###########################################################
    # Q3-2 #
    # Write your own code #
    while head.next:
        chg = False
        dummy = LinkedNode(0)
        dummy.next = head
        while head.next:
            # print("step")
            if head.val == head.next.val:
                chg |= True
                if head.val == 32:
                    return True
                head.val *= 2
                head.next = head.next.next
            else:

                head = head.next
        if not chg:
            return False

        head = dummy.next
    return False
    ###########################################################
```
