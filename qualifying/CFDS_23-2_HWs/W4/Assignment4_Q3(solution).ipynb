```python
#Do not modify this code
class LinkedNode():
  def __init__(self, x:int) -> None:
    self.val = x
    self.next = None
```

```python
def MakeNode(arr:list, root:LinkedNode) -> LinkedNode:
    ###########################################################
    # Q3-1 #
    # Write your own code #
    for i in arr:
        curNode = root
        while curNode.next:
            curNode = curNode.next
        curNode.next = LinkedNode(i)
    ###########################################################
```

```python
def MergeGame(arr:list) -> bool:
    root = LinkedNode(0)
    MakeNode(arr, root)
    head = root.next
    ###########################################################
    # Q3-2 #
    # Write your own code #
    while head:
        if head.next == None:
            break
        if head.val == head.next.val:
            head.val += head.next.val
            head.next = head.next.next
            head = root.next
        else:
            head = head.next
    head = root.next
    while head:
        if head.val == 64:
            return True
        head = head.next
    return False
    ###########################################################
```

```python
def ZeroSumGame(arr:list) -> bool:
    root = LinkedNode(0)
    MakeNode(arr, root)
    head = root.next
    answer = 1
    temp = LinkedNode(0)
    temp.next = head
    ###########################################################
    # Q3-3 #
    # Write your own code #
    hash = {0: temp}
    prev = 0
    while head:
        prev += head.val
        hash[prev] = head
        head = head.next
    head = temp
    prev = 0
    while head:
        prev += head.val
        head.next = hash[prev].next
        head = head.next
    ###########################################################
    compute = temp.next
    while compute:
        answer *= compute.val
        compute = compute.next
    return answer
```
