```python
from collections import deque

class Office():
    # Do not modify this code
    def __init__(self, M, N, H, employees):
        self.M = M  # 건물의 가로 길이
        self.N = N  # 건물의 세로 길이
        self.H = H  # 건물의 높이
        self.offices = [[[0] * M for _ in range(N)] for _ in range(H)]
        self.coordinates_employees = employees # 직원이 소유한 개인사무실의 좌표를 저장하는 리스트
        self.employees()

    # Do not modify this code
    def employees(self):
        for x, y, z in self.coordinates_employees:
            if 0 <= x < self.M and 0 <= y < self.N and 0 <= z < self.H:
                self.offices[z][y][x] = 1

    def initial_stealthguard(self, coordinates_safeguards) :
        ###################################################
        # Q2-1#
        # Write your own code #
        for x, y ,z in coordinates_safeguards:
            if self.offices[z][y][x] == 1:
                self.offices[z][y][x] = 2
        # 사람이 있으면 1
        # 사럼이 없으면 0
        # 사람도 있고 설치도 되어 있으면 2
        
        ###################################################

    def install_stealthguard(self):
        dx = [0, 0, 1, -1, 0, 0]
        dy = [1, -1, 0, 0, 0, 0]
        dz = [0, 0, 0, 0, 1, -1]

        def bfs_or_dfs() :
            hours_needed = 0
            ###################################################
            # Q2-2#
            # Write your own code #
            # 먼저 사람 있고 설치된 곳 찾기 
            for x, y ,z in self.coordinates_employees:
                if self.offices[z][y][x] == 2:
                    start = 0    
                    queue = [[x, y ,z]]

                    while queue:
                        x, y, z = queue.pop(0)
                        print(x, y ,z)
                        self.offices[z][y][x] == start

                        for dxs, dys, dzs in zip(dx, dy, dz):
                            
                            # 이동이 valid하면서, 그 office에 직원이 있으면서 아직 설치가 안된 경우
                            if x+dxs in range(self.M) and y+dys in range(self.N) and z+dzs in range(self.H) and self.offices[z+dzs][y+dys][x+dxs] == 1:
                                queue.append([x+dxs, y+dys, z+dzs])
                        
                        hours_needed += 12

            ###################################################
            return hours_needed


        ###################################################
        # Q2-3 #
        # Write your own code #
        days_needed = bfs_or_dfs()/24

        unguarded_employees = 0
        for x in range(self.M):
            for y in range(self.N):
                for z in range(self.H):
                    if self.offices[z][y][x] == 1:
                        unguarded_employees += 1
        

        ###################################################

        if unguarded_employees > 0 :
            print(f"직원 {unguarded_employees}명의 PC에 SafeGuard 프로그램을 설치할 수 없습니다.")
        else :
            print(f"{days_needed}일 후면, 모든 사무실에 SafeGuard 프로그램이 설치됩니다.")
```
