```python
#Do not modify this code
class MazeNode(): # Tree Node
  def __init__(self, x:list, y:int) -> None:
    self.loc = x
    self.value = y
    self.left = None
    self.right = None
    self.up = None
    self.down = None
```

```python
def MakeNode(node:MazeNode, matrix:list, visited:list) -> None:
    ###########################################################
    # Q1-1 #
    # Write your own code #
    row = node.loc[0]
    col = node.loc[1]
    if col-1 >= 0:
        if matrix[row][col-1] != 0 and visited[row][col-1] == 0: #left
            node.left = MazeNode([row, col-1], matrix[row][col-1])
            visited[row][col-1] = 1
            MakeNode(node.left, matrix, visited)
    if col+1 < len(matrix[0]):
        if matrix[row][col+1] != 0 and visited[row][col+1] == 0: #right
            node.right = MazeNode([row, col+1], matrix[row][col+1])
            visited[row][col+1] = 1
            MakeNode(node.right, matrix, visited)
    if row-1 >= 0:
        if matrix[row-1][col] != 0 and visited[row-1][col] == 0: #up
            node.up = MazeNode([row-1, col], matrix[row-1][col])
            visited[row-1][col] = 1
            MakeNode(node.up, matrix, visited)
    if row+1 < len(matrix):
        if matrix[row+1][col] != 0 and visited[row+1][col] == 0: #down
            node.down = MazeNode([row+1, col], matrix[row+1][col])
            visited[row+1][col] = 1
            MakeNode(node.down, matrix, visited)
    ###########################################################
```

```python
class Maze():
  def __init__(self, root:MazeNode) -> None:
    self.root = root
    self.len = 0
    self.answer = 0

  def print(self):
    print(self.answer)

  def Traversal(self):
  ###########################################################
  # Q1-2 #
  # Write your own code #
    self.Traversalhelp(self.root)

  def Traversalhelp(self, curNode:MazeNode):
    if curNode.value == 3:
      self.answer = self.len
    self.len += 1
    if curNode.left:
      self.Traversalhelp(curNode.left)
    if curNode.right:
      self.Traversalhelp(curNode.right)
    if curNode.up:
      self.Traversalhelp(curNode.up)
    if curNode.down:
      self.Traversalhelp(curNode.down)
    self.len -= 1
  ###########################################################
```

```python
def FindMaze(matrix:list) -> int:
    ###########################################################
    # Q1-3 #
    # Write your own code #
    for i in range(len(matrix)):
        if 2 in matrix[i]:
            m = i
            n = matrix[i].index(2)
    maze = MazeNode([m, n], 2)
    visited = [[0 for j in range(len(matrix[0]))] for k in range(len(matrix))]
    visited[m][n] = 1
    ###########################################################

    #Do not modify this code
    MakeNode(maze, matrix, visited)
    mymaze = Maze(maze)
    mymaze.Traversal()
    mymaze.print()
```
