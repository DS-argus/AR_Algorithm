```python
def t_conquer(x,y,N):
    global white, red
    color = territory[x][y]
    for i in range(x, x+N):
        for j in range(y, y+N):
            if color != territory[i][j]:
                t_conquer(x,y,N//2)
                t_conquer(x+(N//2),y,N//2)
                t_conquer(x,y+(N//2),N//2)
                t_conquer(x+(N//2),y+(N//2),N//2)
                return
    if color == 0:
        white += 1
    else:
        red += 1
```
