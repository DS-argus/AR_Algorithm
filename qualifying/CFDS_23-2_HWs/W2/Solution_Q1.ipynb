```python
def findLength(candles: list, n: int) -> int:
    start, end = 0, max(candles)

    while (start <= end):
        total = 0
        length = (start + end) // 2
        for candle in candles:
            if candle > length:
                total += candle - length

        if total == n :
            return length

        elif total < n :
            end = length - 1

        else:
            start = length + 1

    return -1
```
