```python
# Book Class Definition
# Do not modify this code
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
```

```python
# Library Class Definition

class Library:
    def __init__(self, name):
        self.name = name
        self.books = []

    def add_book(self, book, manager):
        self.books.append(book)
        manager.add_book_to_library(book, self)  # Easy to manage in Library Manager Class

    def show_books(self):
        for i in range(len(self.books)):
            print(self.books[i].title, self.books[i].author, self.books[i].year)

    def sort_books_title(self): #Using Insertion Sort

        for i in range(1, len(self.books)):
            #################################################
            #Q3-1. using one of sorting algorithm, write your own code
            j = i
            while j>=1:
                if self.books[j-1].title > self.books[j].title:
                    self.books[j-1], self.books[j] = self.books[j], self.books[j-1]
                else:
                    break
                j -= 1
            ###################################################

    def search_book(self, query):
        found_books = []
        ###################################################
        # Q3-2. Searching book title or author
        # Partial matches should be searchable
        query = query.lower()
        for book_info in self.books:
            title, author = book_info.title.lower(), book_info.author.lower()
            if query in title or query in author:
                found_books.append(book_info)
        ###################################################
        return found_books
```

```python
# Library Manager Definition
# Adding and Selecting Library
# Searching books from all of libraries

class LibraryManager:
    def __init__(self):
        self.libraries = []
        self.book_to_library_map = {}

    def add_library(self, library):
        self.libraries.append(library)

    def select_library(self, library_name):
        for library in self.libraries:
            if library.name == library_name:
                return library
        return None

    def add_book_to_library(self, book, library):
        self.book_to_library_map[book.title] = library

    def find_library_of_book(self, book_title):
        matching_books = {}
        ###################################################
        # Q3-3. Searching book title from all of libraries
        # Partial matches should be searchable
        for library in self.libraries:
            searched_books = library.search_book(book_title)
            for searched_book in searched_books:
                matching_books[searched_book.title] = library
        ###################################################
        return matching_books
```

```python
with open("library.txt", "r") as file :
    books = []
######################################################
    
# Q3-4. Input library.txt
    while True:
        line = file.readline()
        if not line:
            break
        books.append(line[:-1].split(', '))
######################################################
```

```python
# Classifying Books by Library

# Initial Library Manager
manager = LibraryManager()

# Initial Library
library1 = Library("Main Library")
library2 = Library("Data Science Library")
library3 = Library("Engineering Library")

# Adding Library to Library Manager
manager.add_library(library1)
manager.add_library(library2)
manager.add_library(library3)

for i in range(len(books)):
    books[i][2] = int(books[i][2])
    book = Book(books[i][0], books[i][1], books[i][2])
    if books[i][3] == "Main Library" :
        library1.add_book(book, manager)
    elif books[i][3] == "Data Science Library" :
        library2.add_book(book, manager)
    else :
        library3.add_book(book, manager)
```

```python
library1.sort_books_title()
library2.sort_books_title()
library3.sort_books_title()
```
