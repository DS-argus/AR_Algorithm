```python
import sys
from datetime import datetime, timedelta

def main(class_info_file, class_students_data_file):
    with open(class_info_file, "r") as file:
        class_info = file.read().split()

    ###################################################
    # Q2-1#
    # Write your own code #
    class_name = class_info[0]
    class_begin_time = class_info[1]
    class_end_time = class_info[2]
    student_number = class_info[3]
    ###################################################

    attendance_record(class_name, class_begin_time, class_end_time, student_number, class_students_data_file)


def attendance_record(class_name, class_begin_time, class_end_time, student_number, class_students_data_file):
    time_format = "%H:%M"
    class_begin_time_format = datetime.strptime(class_begin_time, time_format)
    class_end_time_format = datetime.strptime(class_end_time, time_format)
    class_duration = class_end_time_format - class_begin_time_format

    students= {} # dictionary initialion
    count = 0   # 강의실에 들어온 학생 수 initializion

    with open(class_students_data_file, "r") as file:
        while True:
            line = file.readline().strip()
            if not line:
                break

            enter_time, leave_time = [datetime.strptime(time_each, time_format) for time_each in line.split()]

            ###################################################
            # Q2-2#
            # Write your own code #
            # hint : datetime.timedelta() method
            leave_time = min(leave_time, class_end_time_format)
            attendance_duration = leave_time - max(enter_time, class_begin_time_format)
            cond_a = attendance_duration >= 0.9*class_duration
            cond_b = enter_time <= class_begin_time_format + timedelta(hours=1)/6
            cond_c = enter_time > class_begin_time_format + timedelta(hours=1)/6
            cond_d = attendance_duration > 0.5*class_duration and attendance_duration < 0.9*class_duration
            cond_e = attendance_duration <= 0.5*class_duration

            students[count] = [cond_a, cond_b, cond_c, cond_d, cond_e, enter_time, leave_time, attendance_duration/class_duration]

            count += 1

            ###################################################

    attendance = 0 # 출석 학생 수 initialization
    tardiness = 0  # 지각 및 조퇴 학생 수 initialization
    absence = 0    # 결석 학생 수 initialization

    ###################################################
    # Q2.3#
    # Write your own code #
    # hint : 조건문의 처리순서에 유의
    for conds in students.values():
        if conds[4]:
            absence += 1
        elif conds[2] or conds[3]:
            tardiness += 1
        elif conds[0] and conds[1]:
            attendance += 1

    absence += int(student_number) - count
    

    ###################################################

    print(f"{class_name} 과목의 출결현황입니다. 총수강생: {student_number}명 中 출석: {attendance}명, 지각 및 조퇴: {tardiness}명, 결석: {absence}명")
```
