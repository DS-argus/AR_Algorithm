```python
#Q3-1 Encrypt text when the text is entered
def encryption(text: str) -> str:
  crypto = ""
  ###################################################
  # Q3-1 #
  # Write your own code #
  if (len(text) + 4)/4 == int((len(text) + 4)/4):
    k = int((len(text) + 4)/4)
  else:
    k = int((len(text) + 4)/4)+1
  
  n = 2*k-1
  store = []

  i = 0
  while i < len(text):
    if i<n:
      store.append([text[i]])
      i += 1
    else:
      store[n-2-i].insert(0, text[i])
      i += 1

  for l in store:
    for j in l:
      crypto += j  
  ###################################################
  print(crypto)
  return crypto
```

```python
#Q3-2 Decrypt text when the text is entered
def decryption(text: str) -> None:
  crypto = ""
  ###################################################
  # Q3-2 #
  # Write your own code #
  if (len(text) + 4)/4 == int((len(text) + 4)/4):
    k = int((len(text) + 4)/4)
  else:
    k = int((len(text) + 4)/4)+1
  
  n = 2*k-1

  full_size = 4*k-4
  left = full_size-len(text)
  store = []
  
  i = 0
  while i < len(text):
    if i == 0 or i == len(text)-1:
      store.append([text[i]])
    elif i<=left:
      store.append([0, text[i]])
    else:
      store.append([text[i], text[i+1]])
      i += 1
    
    i += 1

  i = 0
  while i < len(text):
    if i<n:
      crypto += store[i][-1]
      i += 1
    else:
      crypto += store[n-2-i][0]
      i += 1

  ###################################################
  print(crypto)
```
