```python
#Do not modify this code
class GSDS(): # GSDS Graduate School
    def __init__(self) -> None:
        self.number_of_students = {'ms': 0, 'phd': 0}
        self.students = dict()

    # Enter the number of master's and doctoral students in the GSDS Graduate School
    def studentAccount(self, studentType: str) -> None:
        self.number_of_students[studentType] += 1

    # Update student information, including subject names, credits, and grades (Don't allow retakes)
    def studentUpdate(self, studentName: str, subjectName: str, credit: int, grade: float) -> None:
        if not self.students.get(studentName, False):
          self.students[studentName] = {subjectName:[credit, grade]}
          print("studentName : %s, subjectName : %s, credit : %d, score : %.1f" %(studentName, subjectName, credit, grade))
        else:
          if not self.students[studentName].get(subjectName, False):
            self.students[studentName][subjectName] = [credit, grade]
            print("studentName : %s, subjectName : %s, credit : %d, score : %.1f" %(studentName, subjectName, credit, grade))
          else:
            print("Retakes are not permitted")

    # Call the total number of students in the GSDS Graduate School
    def studentTotal(self) -> None:
        print("The total number of students at GSDS graduate school is %d ms and %d phd." %(self.number_of_students['ms'], self.number_of_students['phd']))
```

```python
#Do not modify this code
class GSDS_2022(): # GSDS Graduate requirements for 2022
    def __init__(self, grad: GSDS) -> None:
        self.grad = grad
        self.requiredcredit = {'ms': 12, 'phd': 24}
        self.requiredavggrade = 3.0
```

```python
#Do not modify this code
class GSDS_2023(): # GSDS Graduate requirements for 2023
    def __init__(self, grad: GSDS) -> None:
        self.grad = grad
        self.requiredcredit = {'ms': 15, 'phd': 27}
        self.requiredavggrade = 3.3
```

```python
class Student(): # GSDS Graduate Student
    #Do not modify this code
    def __init__(self, grad_year, name: str, studentType: str):
        self.year = grad_year
        self.name = name
        self.studentType = studentType
        self.year.grad.studentAccount(studentType)

    # Q1-1
    # Update the student's name and course information, including subject name, credits, and grade
    def updateSubject(self, subjectName: str, credit: int, grade: float) -> None:
        ###################################################
        # Q1-1 #
        # Write your own code #
        self.year.grad.studentUpdate(self.name, subjectName, credit, grade)

        ###################################################
        pass

    # Q1-2
    # Check credits and average grade the student have taken to date
    def checkSubject(self) -> None:
        self.total_credit = 0
        self.avg_grade = 0
        total_grade = 0
        ###################################################
        # Q1-2 #
        # Write your own code #

        for key in self.year.grad.students[self.name]:
            self.total_credit += self.year.grad.students[self.name][key][0]
            total_grade += self.year.grad.students[self.name][key][1]*self.year.grad.students[self.name][key][0]

        self.avg_grade = total_grade / self.total_credit

        ###################################################
        print("Total Credit : %d, Avg Grade : %.1f" %(self.total_credit, self.avg_grade))
        pass

    # Q1-3
    # Check the student have met graduation requirements
    def checkGraduation(self) -> None:
        ###################################################
        # Q1-3 #
        # Write your own code #
        if self.avg_grade >= self.year.requiredavggrade and \
            self.total_credit >= self.year.requiredcredit[self.studentType]:
        ###################################################
          print("Graduation requirements are met.")
        else:
          print("Graduation requirements are not met.")
        pass
```
