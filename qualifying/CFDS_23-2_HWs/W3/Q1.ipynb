# Assignment 3.1 Naive Word Processor

```python
from collections import deque

class Naive_WordProcessor:
    def __init__(self):
      ########Do not modify########
      self.stack = []
      self.queue = deque([])
      ############################

    def addText(self, text: str) -> str:
      ########Write your own code#########
      for letter in text:
        self.stack.append(letter)
      ####################################
      return "".join(self.stack) + '|' + "".join(self.queue)

    def deleteText(self, k: int) -> str:
      ########Write your own code#########
      for _ in range(k):
        self.stack.pop()
        if len(self.stack) == 0:
          break
      ####################################
      return "".join(self.stack) + '|' + "".join(self.queue)

    def cursorLeft(self, k: int) -> str:
      ########Write your own code#########
      for _ in range(k):
        popped = self.stack.pop()
        self.queue.appendleft(popped)
        if len(self.stack) == 0:
          break
      ####################################
      return "".join(self.stack) + '|' + "".join(self.queue)

    def cursorRight(self, k: int) -> str:
      ########Write your own code#########
      for _ in range(k):
        popped = self.queue.popleft()
        self.stack.append(popped)
        if len(self.queue) == 0:
          break
      ####################################
      return "".join(self.stack) + '|' + "".join(self.queue)
```

```python
# Given Example 1
TE = Naive_WordProcessor()
print(TE.addText('I love python!'))
print(TE.deleteText(5))
print(TE.addText('practice'))
print(TE.cursorLeft(8))
print(TE.cursorRight(10))
print(TE.deleteText(10))
```
